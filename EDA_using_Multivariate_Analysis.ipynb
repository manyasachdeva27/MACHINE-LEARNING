```python
#1. Scatterplot (Numerical -numerical
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',style='smoker',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',style='smoker',size='size',data=tips)
```

```python
#ii) Barplot
sns.barplot(x='Pclass',y='Age',data=titanic)
```

```python
sns.barplot(x='Pclass',y='Fare',data=titanic)
```

```python
sns.barplot(x='Pclass',y='Fare',hue='Sex',data=titanic)
```

```python
sns.barplot(x='Pclass',y='Age',hue='Sex',data=titanic)
```

```python
#iii)Boxplot
sns.boxplot(x='Sex',y='Age',data=titanic)
```

```python
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)
```

```python
#iv)Distplot
sns.distplot(titanic['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
```

```python
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
#v)Heatmap
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Embarked')['Survived'].mean()*100
```

```python
#vi) Clusterplot
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

```python
#vii)Pairplot
sns.pairplot(iris,hue='species')
```

```python
#viii) Lineplot
sns.lineplot(x='year',y='passengers',data=flights)
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
