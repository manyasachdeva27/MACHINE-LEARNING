## Verify Persistent Blocking

### Subtask:
Attempt another request to `https://www.ambitionbox.com/list-of-companies?page=1` with a slightly modified User-Agent to confirm the blocking is consistent and not a transient issue or specific header detection.


**Reasoning**:
To verify persistent blocking, I will define a new `User-Agent` string, subtly different from the previous one, and then make a `requests.get()` call to the target URL with this new `User-Agent`.



```python
new_headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) AppleWebKit/537.36 (KHTML , like Gecko) Chrome/81.0.40'}
webpage_new = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=new_headers)
print(webpage_new.status_code)
print(webpage_new.text[:500]) # Print first 500 characters to avoid flooding output
```
