```python
#opening a csv file from a URL

import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers) #requets.get to fetch all the content form the website
data = StringIO(req.text) #req.text to fetch all the text only form the website

df = pd.read_csv(data)
df
```

```python
#sep parameter for tsv files use pd.read_csv('/content/city_temperature.tsv',sep='/t')
```

```python
#index_col paramter
pd.read_csv('/content/DailyDelhiClimateTrain.csv',index_col='date') #index_col aisa colum hoga jo aapke analysis me kaam nhi aane vala just for unique id
```

```python
#Header parameter
#jab humne first row ko column ke names me convert hota h
pd.read_csv('/content/DailyDelhiClimateTrain.csv',header=1) #header=1 aapka jo first row h vo 1 se start hho rahi h
```

```python
#use_cols parameter
pd.read_csv('/content/DailyDelhiClimateTrain.csv',usecols=['date','wind_speed','meantemp'])
```

```python
#skip rows/ nrows parameter
pd.read_csv('/content/DailyDelhiClimateTrain.csv',skiprows=[1,2])
```

```python
pd.read_csv('/content/DailyDelhiClimateTrain.csv',nrows=2)
```

```python
#Encoding parameter used for datasets not encoded with utf-8 like datasets like hindi datasets, emoji datasets
pd.read_csv('/content/DailyDelhiClimateTrain.csv',encoding='latin-1')
```

```python
#Skip bad lines (parser error)
#kuch aise datasets jinme beech beech me kuch rows me gadbad aayega jaise sab rows me 5 values but ek do rows me 6 values to vo read nahi kar pata h pandas
pd.read_csv('/content/DailyDelhiClimateTrain.csv',on_bad_lines='skip')
```

```python
from numpy import int64
#dtypes parameter
pd.read_csv('/content/DailyDelhiClimateTrain.csv').info()
```

```python
pd.read_csv('/content/DailyDelhiClimateTrain.csv',dtype={'Day':int})
```

```python
#Handling Dates
#dates are always passed in stripg (object format) so we have to convert them in date time format
pd.read_csv('/content/DailyDelhiClimateTrain.csv',parse_dates=['date'])
```

```python
#Convertors
pd.read_csv('/content/DailyDelhiClimateTrain.csv',converters={'meantemp':lambda x:float(x)/100})
```

```python
#na_values parameter
pd.read_csv('/content/DailyDelhiClimateTrain.csv',na_values=['Nan','nan','na-n'])
```

```python
#Loading a huge dataset in chunks
dffs=pd.read_csv('/content/DailyDelhiClimateTrain.csv',chunksize=10000)
#ab aapko jo bhi kaam karna h ape dataframe ske upar aap loop me karoge
```

```python
for chunk in dffs:
  print(chunk)
```
