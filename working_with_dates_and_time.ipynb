# Working with Dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

#### 1. Extract year

```python
date['date_year'] = date['date'].dt.year

date.sample(5)
```

#### 2. Extract Month

```python
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

#### Extract Days

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

#### Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

#### Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

#### Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

#### Extract Time elapsed between dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

# The unit 'M' for months is not supported by np.timedelta64 because months have varying lengths.
# We can approximate the number of months by dividing the total number of days by the average number of days in a month.
# An average month has approximately 30.44 days (365.25 days / 12 months).
np.round((today - date['date']).dt.days / 30.44, 0)
```

```python
# Converting to datetime datatype
time['date'] = pd.to_datetime(time['date'])
```

```python
time.info()
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

#### Extract Time part

```python
time['time'] = time['date'].dt.time

time.head()
```

#### Time difference

```python
today - time['date']
```

```python
# in seconds

(today - time['date'])/np.timedelta64(1,'s')
```

```python
# in minutes

(today - time['date'])/np.timedelta64(1,'m')
```

```python
# in hours

(today - time['date'])/np.timedelta64(1,'h')
```
