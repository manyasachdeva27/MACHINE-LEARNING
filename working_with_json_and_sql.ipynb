```python
import sqlite3

# create database
conn = sqlite3.connect("world.db")
cursor = conn.cursor()

# Define table creation SQL for SQLite
create_table_sql = """
CREATE TABLE IF NOT EXISTS country (
    Code TEXT PRIMARY KEY,
    Name TEXT,
    IndepYear INTEGER
);

CREATE TABLE IF NOT EXISTS city (
    ID INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT,
    CountryCode TEXT,
    Population INTEGER,
    FOREIGN KEY (CountryCode) REFERENCES country(Code)
);

CREATE TABLE IF NOT EXISTS countrylanguage (
    CountryCode TEXT,
    Language TEXT,
    IsOfficial TEXT,
    Percentage REAL,
    PRIMARY KEY (CountryCode, Language),
    FOREIGN KEY (CountryCode) REFERENCES country(Code)
);
"""

# Execute table creation SQL
cursor.executescript(create_table_sql)

# read remaining sql file (queries)
with open("world.sql", "r", encoding="utf-8") as f:
    sql_script = f.read()

# Remove 'use world;' if it exists, as it's not valid for SQLite
sql_script = sql_script.replace("use world;\n", "")

# Remove MySQL-specific backticks, as they are not valid for SQLite
sql_script = sql_script.replace("`", "")

# Correct specific malformed SQL: 'PRIMARYselect' to 'select'
sql_script = sql_script.replace("PRIMARYselect", "select")

# Remove lines starting with '#' which are not valid comments in SQLite
sql_script_lines = sql_script.split('\n')
sql_script = '\n'.join([line for line in sql_script_lines if not line.strip().startswith('#')])

# execute script (now tables should exist)
cursor.executescript(sql_script)

conn.commit()

print("Database Loaded Successfully!")
```
